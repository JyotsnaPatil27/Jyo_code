```
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit as cf
```

```
Pul1 = open("/content/drive/My Drive/Pulse1.txt","r").readlines()
Pul11 = []
for x in Pul1:
  Pul11.append(float(x.split(' ')[0].strip()))
Pul12 = np.array(Pul11)
x = np.arange(540,630,1)
plt.plot(x,Pul12)
```

```
def gaussian(x,mu,sigma,a): 
    return a*np.exp(-1*((x-mu)**2)/(sigma**2))
```

```
x = x
y = Pul12
mu = np.argmax(Pul12)
mean = np.mean(Pul12)
sigma = np.std(Pul12)
p0 = mu, mean,sigma
```

```
p_opt, p_cov = cf(gaussian, x, y, p0)
plt.plot(x,Pul12)
plt.plot(x, gaussian(x,  *p_opt), color = "red", label = "Gaussian Fit")
```
